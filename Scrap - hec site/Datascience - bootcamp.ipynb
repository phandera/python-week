```python
import pandas as pd
import html5lib
from bs4 import BeautifulSoup
import requests
```

```python
url="https://hec.ac.mw/scholarships"
```

```python
response = requests.get(url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html5lib')
print(soup.prettify())
```

```python
#block = soup.find('p', class_ = 'scholar-meta')
#block
```

```python
blocks = soup.find_all('div', class_ = 'scholar-content-box')
block2 = soup.find('div', class_ = 'scholar-content-box').find_all('p')
block2[2].text
#block.text

```

```python

dates=[]
scholarships=[]
close_date=[]
for block in blocks :
    dates.append(block.find('p', {'class':'scholar-meta'}).text.replace('Posted on ','').replace(' by HEC Media','').strip())
    scholarships.append(block.find('h4', {'class':'scholarship-title'}).text.strip())
    close_date.append(block.find_all('p')[2].text.strip().split("Closes ")[1])
    
    df=pd.DataFrame({"Posted date":dates,"Scholarship":scholarships,"Deadline":close_date})
    
```
